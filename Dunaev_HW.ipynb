Для непрерывного начисления необходимо задать параметр period равным 1000

По дефолту капитализация - 1 раз в год

```python
import numpy as np

def npv(nominal, is_compounding, rate, term,discount,period=1):
    if is_compounding:
        if period != 1000:
            return nominal*((1+rate/period)**(term*period)/((1+discount)**term)-1)
        else:
            return nominal*(np.exp(rate*term)/((1+discount)**term)-1)
    else:
        npv_ = term*nominal*((1+rate)/((1+discount))-1) 
        return npv_
```

```python
print('номинал 1000''\n'
      'без капитализации''\n'
      'купонная доходность 3%''\n'
      'срок - 4.3 года''\n'
      'ставка дисконтирования - 2.8%''\n'
      ,'NPV = ', npv(1000, 0, 0.03, 4.3, 0.028))
print('\n''\n''\n')
print('номинал 1000''\n'
      'без капитализации''\n'
      'купонная доходность 23%''\n'
      'срок - 0.25 года''\n'
      'ставка дисконтирования - 19%''\n'
      ,'NPV = ', npv(1000, 0, 0.23, 0.25, 0.19))
print('\n''\n''\n')
print('номинал 1000''\n'
      'без капитализации''\n'
      'купонная доходность 5%''\n'
      'срок - 1.3 года''\n'
      'ставка дисконтирования - 2%''\n'
      ,'NPV = ', npv(1000, 0, 0.05, 1.3, 0.02))
print('\n''\n''\n')
print('номинал 1000''\n'
      'c капитализацией''\n'
      'купонная доходность 5%''\n'
      'срок - 1.3 года''\n'
      'ставка дисконтирования - 2%''\n'
      'периодичность начисления: 3 раза в год''\n'
      ,'NPV = ', npv(1000, 1, 0.05, 1.3, 0.02, 3))
print('\n''\n''\n')
print('номинал 1000''\n'
      'c капитализацией''\n'
      'купонная доходность 5%''\n'
      'срок - 1.3 года''\n'
      'ставка дисконтирования - 2%''\n'
      'периодичность начисления: 365 раз в год''\n'
      ,'NPV = ', npv(1000, 1, 0.05, 1.3, 0.02, 365))
print('\n''\n''\n')
print('номинал 1000''\n'
      'c капитализацией''\n'
      'купонная доходность 5%''\n'
      'срок - 1.3 года''\n'
      'ставка дисконтирования - 2%''\n'
      'периодичность начисления:  непрерывно''\n'
      ,'NPV = ', npv(1000, 1, 0.05, 1.3, 0.02, 1000))
```

```python
def irr(nominal, is_compounding, rate, term,period=1):
    if is_compounding:
        if period != 1000:
            irr_=((1+rate/period)**(term*period))**(1/term) - 1
        else:
            irr_ = (np.exp(rate*term))**(1/term) - 1 
    else:
        irr_ = rate
    return irr_
```

```python
print('номинал 1000''\n'
      'без капитализации''\n'
      'купонная доходность 3%''\n'
      'срок - 4.3 года''\n'
      ,'IRR = ', irr(1000, 0, 0.03, 4.3)*100,'%')
print('\n''\n''\n')
print('номинал 1000''\n'
      'без капитализации''\n'
      'купонная доходность 5%''\n'
      'срок - 1.3 года''\n'
      ,'IRR = ', irr(1000, 0, 0.05, 1.3)*100,'%')
print('\n''\n''\n')
print('номинал 1000''\n'
      'c капитализацией''\n'
      'купонная доходность 5%''\n'
      'срок - 1.3 года''\n'
      'периодичность начисления: 3 раза в год''\n'
      ,'IRR = ', irr(1000, 1, 0.05, 1.3, 3)*100,'%')
print('\n''\n''\n')
print('номинал 1000''\n'
      'c капитализацией''\n'
      'купонная доходность 5%''\n'
      'срок - 1.3 года''\n'
      'периодичность начисления: 365 раз в год''\n'
      ,'IRR = ', irr(1000, 1, 0.05, 1.3, 365)*100,'%')
print('\n''\n''\n')
print('номинал 1000''\n'
      'c капитализацией''\n'
      'купонная доходность 5%''\n'
      'срок - 1.3 года''\n'
      'периодичность начисления:  непрерывно''\n'
      ,'IRR = ', irr(1000, 1, 0.05, 1.3, 1000)*100,'%')
```
